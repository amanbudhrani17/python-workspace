```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import datasets
```

```python
x,y= datasets.make_classification(n_samples=1000, n_features=5, n_classes=2, n_redundant=0, n_clusters_per_class=1, weights=[0.5,0.3], random_state=42)
fig,ax= plt.subplots(figsize=(9,6))
plt.xlabel("X0", fontsize=15)
plt.ylabel("X1", fontsize=15)
plt.scatter(x[:,0], x[:,1], s=50, c=y)
```

```python
labels = np.array([0,1]*500)
labels = labels.reshape(1000,1)
np.random.seed(42)
weights = np.random.rand(5,1)
bias = np.random.rand(1)
lr = 0.01
def sigmoid(x):
    return 1/(1+np.exp(-x))
def sigmoid_der(x):
    return sigmoid(x)*(1-sigmoid(x))
```

```python
for epoch in range(8000):
    inputs = x

    # feedforward step1
    XW = np.dot(x, weights) + bias

    #feedforward step2
    z = sigmoid(XW)


    # backpropagation step 1
    error = z - labels

    print(error.sum())

    # backpropagation step 2
    dcost_dpred = error
    dpred_dz = sigmoid_der(z)

    z_delta = dcost_dpred * dpred_dz

    inputs = x.T
    weights -= lr * np.dot(inputs, z_delta)

    for num in z_delta:
        bias -= lr * num
```
